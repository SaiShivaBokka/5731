Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):
import nltk

from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rq
import matplotlib.pyplot as plt
import nltk
from wordcloud import WordCloud
from nltk.corpus import stopwords

link = 'https://www.imdb.com/title/tt0468569/reviews/?ref_=tt_ql_urv'
data_frame = pd.DataFrame(columns=['Views'])
request = rq.get(link).text
soup = bs(request,'html.parser')
reviews_txt = soup.find_all('div', attrs={'class' : 'text show-more__control'})
views = []
for i in range(len(reviews_txt)):
    views.append(reviews_txt[i].get_text())  
data_frame['Views'] = views

data_frame['Counts'] = data_frame['Views'].apply(lambda x: len(str(x).split(" ")))
data_frame[['Views','Counts']].head()
```

```python
data_frame['Counts'] = data_frame['Views'].str.len()
data_frame[['Views','Counts']].head()
```

```python
from nltk.corpus import stopwords

import nltk
nltk.download('stops')
No = stopwords.words('english')

data_frame['stops'] = data_frame['Views'].apply(lambda x: len([x for x in x.split() if x in No]))
data_frame[['Views','stops']].head()
```

```python
data_frame['hashs'] = data_frame['Views'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
data_frame[['Views','hashs']].head()
```

```python
data_frame['nums'] = data_frame['Views'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
data_frame[['Views','nums']].head()
```

```python
data_frame['up'] = data_frame['Views'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
data_frame[['Views','up']].head()
```
