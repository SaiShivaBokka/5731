```python
# Write your code here

import pandas as pd

df = pd.read_csv("papers.csv")
df.head()
```

```python
df = df.drop(columns=['id', 'event_type', 'pdf_name'], axis=1).sample(100)
df.head()
```

```python
import re
df['processed_text'] = \
df['paper_text'].map(lambda x: re.sub('[,\.!?]', '', x))
df['processed_text'] = \
df['processed_text'].map(lambda x: x.lower())
df['processed_text'].head()
```

```python
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_word = stopwords.words('english')
stop_word.extend(['from', 'subject', 're', 'edu', 'use'])
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_word] for doc in texts]
data = df.processed_text.values.tolist()
dataword = list(sent_to_words(data))
# remove stop words
dataword = remove_stopwords(dataword)
print(dataword[:1][0][:30])
```

```python
import gensim.corpora as corpora

id2word = corpora.Dictionary(dataword)

texts = dataword

corpus = [id2word.doc2bow(text) for text in texts]

print(corpus[:1][0][:30])
```

```python
from pprint import pprint

num_topics = 10

lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)

pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

## (2) (10 points) Generate K topics by using LSA, the number of topics K should be decided by the coherence score, then summarize what are the topics. You may refer the code here:

https://www.datacamp.com/community/tutorials/discovering-hidden-topics-python

```python
# Write your code here

import pandas as pd

df = pd.read_csv('Musical_instruments_reviews.csv', usecols=['reviewerID', 'reviewText'])
df.head()
```

```python
print(df.shape)
df = df.dropna()
print(df.shape)
```

```python
from gensim.parsing.preprocessing import remove_stopwords, strip_punctuation \
                                        , preprocess_string, strip_short, stem_text
                                        
def preprocess(text):
  filtered = [lambda x: x.lower(),
              remove_stopwords,
              strip_punctuation,
              strip_short,
              stem_text]  
  text = preprocess_string(text, filtered)
  return text
df['Text (Clean)'] = df['reviewText'].apply(lambda x: preprocess(x))
```

```python
from gensim import corpora

corpus = df['Text (Clean)']
dicti = corpora.Dictionary(corpus)

bow = [dicti.doc2bow(text) for text in corpus]
```

```python
from gensim.models import LsiModel
from gensim.models.coherencemodel import CoherenceModel

for i in range(2,11):
    lsi = LsiModel(bow, num_topics=i, id2word=dicti)
    coherence_model = CoherenceModel(model=lsi, texts=df['Text (Clean)'], dictionary=dicti, coherence='c_v')
    coherence_score = coherence_model.get_coherence()
    print('Coherence score with {} clusters: {}'.format(i, coherence_score))
```

```python
lsi = LsiModel(bow, num_topics=10, id2word=dicti)

for topic_num, words in lsi.print_topics(num_words=5):
    print('Words in {}: {}.'.format(topic_num, words))
```

## (4) (10 points) Generate K topics by using BERTopic, the number of topics K should be decided by the coherence score, then summarize what are the topics. You may refer the code here: 

https://colab.research.google.com/drive/1FieRA9fLdkQEGDIMYl0I3MCjSUKVF8C-?usp=sharing

```python
# Write your code here
df = pd.read_csv('tokyo_2020_tweets.csv')
df = df.dropna()
df = df.sample(20000, random_state=42)
df.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

stop_words = stopwords.words('english')

def cln(k):
  k = str(k)
  k = k.lower()
  k = re.sub(r'#[A-Za-z0-9]*', ' ', k)
  k = re.sub(r'https*://.*', ' ', k)
  k = re.sub(r'@[A-Za-z0-9]+', ' ', k)
  tokens = word_tokenize(k)
  k = ' '.join([w for w in tokens if not w.lower() in stop_words])
  k = re.sub(r'[%s]' % re.escape('!"#$%&\()*+,-./:;<=>?@[\\]^_`{|}~“…”’'), ' ', k)
  k = re.sub(r'\d+', ' ', k)
  k = re.sub(r'\n+', ' ', k)
  k = re.sub(r'\s{2,}', ' ', k)
  return k


df['cln'] = df.text.apply(cln)
df.head()
```

```python
! pip install -q bertopic
```

```python
from bertopic import BERTopic
tweets = df.cln.to_list()
timestamp = df.date.to_list()

tpc_mdl = BERTopic(language="english")
topics, probs = tpc_mdl.fit_transform(tweets)

tpc_mdl.get_topic_info()
```
